# Function to count symbols and URLs in the corpus

```python
import re
def count_symbols_and_urls(data):
    num_mentions = 0
    num_hashtags = 0
    num_commas = 0
    num_urls = 0
    
    for tweet in data:
        num_mentions += tweet.count('@')
        num_hashtags += tweet.count('#')
        num_commas += tweet.count(',')
#         num_urls += tweet.count('http://') + tweet.count('https://')
        num_urls += len(re.findall(r'(https?://\S+)', tweet))
    
    print("Number of '@' mentions: ", num_mentions)
    print("Number of '#' hashtags:", num_hashtags)
    print("Number of commas: ", num_commas)
    print("Number of URLs: ", num_urls)
```

```python
def remove_symbols_and_urls(corpus):
    cleaned_tweets = []
    
    for tweet in corpus[:100]:
        cleaned_tweet = tweet.replace('@', '').replace('#', '').replace(',', '')
        cleaned_tweet = ' '.join(filter(lambda x: 'http' not in x, cleaned_tweet.split()))
        cleaned_tweets.append(cleaned_tweet)
    
    # Update the original CSV file with the cleaned tweets
    df.loc[:99, 'TweetText'] = cleaned_tweets
    df.to_csv('full-corpus.csv', index=False)

# Load the CSV file
df = pd.read_csv('full-corpus.csv')

# Extract the tweet text column
corpus = df['TweetText']


```

```python
import spacy
import csv

# Function to perform tokenization of tweets
def tokenize_tweets(df):
    nlp = spacy.load('en_core_web_sm')
    tokenized_tweets = []

    for tweet in df[:100]:
        doc = nlp(tweet)
        tokens = [token.text for token in doc]
        tokenized_tweets.append(tokens)

    return tokenized_tweets


```

```python
# Load the CSV file
df = []
with open('full-corpus.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        df.append(row['TweetText'])

```

```python
import spacy
import csv

# Function to perform lemmatization of tweets
def lemmatize_tweets(corpus):
    nlp = spacy.load('en_core_web_sm')
    lemmatized_tweets = []

    for tweet in corpus[:100]:
        doc = nlp(tweet)
        lemmas = [token.lemma_ for token in doc]
        lemmatized_tweets.append(lemmas)

    return lemmatized_tweets

# Load the CSV file
corpus = []
with open('full-corpus.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        corpus.append(row['TweetText'])

# Lemmatize the first 100 tweets
lemmatized_tweets = lemmatize_tweets(corpus)

# Print the lemmatized tweets
for lemmas in lemmatized_tweets:
    print(lemmas)
```
